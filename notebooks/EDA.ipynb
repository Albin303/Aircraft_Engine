```python
import seaborn as sns

# If you have many features, pick the most relevant for quick plotting
features = [
    'RUL', 'Health_State', 'HPT_outlet_temperature_R', 'Fan_stall_margin',
    'Temp_Ratio_HPT_LPC', 'Press_Ratio_Fan', 'Fuel_Flow_per_Speed'
]
sns.pairplot(df[features], diag_kind='kde', hue="Health_State" if "Health_State" in df.columns else None)
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Scatterplot: Feature vs. RUL
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='RUL', y='HPT_outlet_temperature_R', alpha=0.5)
plt.title('HPT Outlet Temperature vs RUL')
plt.show()
```

```python
engine_id_to_plot = df['Engine_ID'].iloc[0]  # or pick a specific ID

df_engine = df[df['Engine_ID'] == engine_id_to_plot]

plt.figure(figsize=(14,6))
plt.plot(df_engine['Flight_cycle_number'], df_engine['HPT_outlet_temperature_R'], label='HPT Out Temp')
plt.plot(df_engine['Flight_cycle_number'], df_engine['HPT_outlet_temperature_R_rolling_mean'], label='HPT Out Temp (Rolling Mean)')
plt.xlabel('Flight Cycle')
plt.ylabel('Temperature')
plt.title(f'High Pressure Turbine Outlet Temp vs Flight Cycle (Engine {engine_id_to_plot})')
plt.legend()
plt.show()
```

```python
for engine in df['Engine_ID'].unique()[:5]:  # Plot first 5 engines for a start
    temp = df[df['Engine_ID'] == engine]
    plt.plot(temp['Flight_cycle_number'], temp['HPT_outlet_temperature_R'])
    plt.title(f'Engine {engine}: HPT Outlet Temp vs Cycle')
    plt.xlabel('Cycle')
    plt.ylabel('Temp')
    plt.show()
```

```python
# Filter for one engine (for example, Engine_ID == 216)
engine_id = 216
df_engine = df[df['Engine_ID'] == engine_id]

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(df_engine['Flight_cycle_number'], df_engine['HPT_outlet_temperature_R'], label='HPT Outlet Temp')
plt.plot(df_engine['Flight_cycle_number'], df_engine['Fan_stall_margin'], label='Fan Stall Margin')
plt.xlabel('Flight Cycle Number')
plt.ylabel('Sensor Value')
plt.title(f'Engine {engine_id} - Key Features over Cycles')
plt.legend()
plt.show()
```

```python
plt.scatter(df_engine['RUL'], df_engine['HPT_outlet_temperature_R'], alpha=0.7)
plt.xlabel('RUL')
plt.ylabel('HPT Outlet Temp')
plt.title(f'Engine {engine_id}: HPT Outlet Temp vs RUL')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_engine['Flight_cycle_number'], df_engine['HPT_outlet_temperature_R'], label='Raw Value')
plt.plot(df_engine['Flight_cycle_number'], df_engine['HPT_outlet_temperature_R_rolling_mean'], label='Rolling Mean', linestyle='--')
plt.xlabel('Flight Cycle Number')
plt.ylabel('HPT Outlet Temp')
plt.title(f'Engine {engine_id}: HPT Outlet Temp and Trend')
plt.legend()
plt.show()
```
